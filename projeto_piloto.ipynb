```python
df_cor = tabela.cor.value_counts().to_frame().reset_index() #to_frame transfroma uma serie em DataFrame, reset_index() para transformar em 2 colunas e grava em 'df_cor'
df_cor.columns = ["cor", "total"] #Renomeia as colunas
df_cor
```

```python
df_cor.loc[df_cor.cor==1, "cor"]='brancos' #Substitui valores da coluna por condição em pandas
df_cor.loc[df_cor.cor==2, "cor"]='negros' #Substitui valores da coluna por condição em pandas
```

```python
import seaborn as sns
sns.barplot(x="cor", y="total", data = df_cor) #Gráfico de barras com seaborn, definindo a "cor" no eixo x e o "total" no eixo y, extraido do DF df_cor
```

```python
motivacao_geral_por_curso = tabela.groupby("cod_curso").sum()["motivacao"] #Agrupa o "cod_curso" e soma a coluna "motivacao" de acordo com o curso na coluna "cod_curso". Quando extraimos uma coluna, é devolvido uma série
```

```python
motivacao_geral_por_curso
```

```python
motivacao_geral_por_curso = motivacao_geral_por_curso.to_frame().reset_index()
```

```python
motivacao_geral_por_curso.columns = ["curso", "motivacao_total"]
motivacao_geral_por_curso
```

```python
plt.figure(figsize=(18,10))
sns.barplot(x="curso", y="motivacao_total", data = motivacao_geral_por_curso)
```
